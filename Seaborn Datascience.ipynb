## Exploring Age Variable

<p>Visualize distribution of Age variable with Seaborn distplot() function¶</p>
<li>
<ul>Seaborn distplot() function flexibly plots a univariate distribution of observations.</ul>
<ul>This function combines the matplotlib hist function (with automatic calculation of a good default bin size) with the seaborn kdeplot() and rugplot() functions.<ul>
<ul>So, let's visualize the distribution of Age variable with Seaborn distplot() function.</ul>
</li>

```python
f,ax=plt.subplots(figsize=(8,6))
x=fifa19data['Age']
ax=sns.distplot(x,bins=10)
plt.show()
```

<p>It can be seen that the Age variable is slighthly positively skewed</p>

<p>We can use Pandas series object to get an informative axis</p>

```python
f,ax =plt.subplots(figsize=(8,6))
x=fifa19data['Age']
x=pd.Series(x,name="Age variable")
ax=sns.distplot(x,bins=10)
plt.show()
```

## plotting the distribution on the vertical axis

```python
f,ax=plt.subplots(figsize=(8,6))
x=fifa19data['Age']
ax=sns.distplot(x,bins=10,vertical=True)
plt.show()
```

### Seaborn Kernel Density Estimation (KDE)Plot

<li>The kernel density estimate (KDE) plot is a useful tool for plotting the shape of a distribution.</li>
<li>Seaborn kdeplot is another seaborn plotting function that fits and plot a univariate or bivariate kernel density estimate.</li>
<li>Like the histogram, the KDE plots encode the density of observations on one axis with height along the other axis.</li>
<li>We can plot a KDE plot as follows-</li>

```python
f,ax=plt.subplots(figsize=(8,6))
x=fifa19data['Age']
x=pd.Series(x,name='Age variable')
ax=sns.kdeplot(x)
plt.show()
```

<p>We can shade under the density curve and use a different color as follows:-</p>

```python
f,ax=plt.subplots(figsize=(8,6))
x=fifa19data['Age']
x=pd.Series(x,name="Age variable")
ax=sns.kdeplot(x,shade=True,color='b')
plt.show()
```

## Histograms

<li>A histogram represents the distribution of data by forming bins along the range of the data and then drawing bars to show the number of observations that fall in each bin.</li>
<li>A hist() function already exists in matplotlib</li>
<li>We can use Seaborn to plot a histogram.</li>

```python
f,ax=plt.subplots(figsize=(8,6))
x=fifa19data['Age']
ax=sns.distplot(x,kde=False,rug=True,bins=10)
plt.show()
```

<p>We can plot a KDE plot alternatively as follows:-</p>

```python
f,ax=plt.subplots(figsize=(8,6))
x=fifa19data['Age']
ax=sns.distplot(x,hist=False,rug=True,bins=10)
plt.show()
```

<h5>Seaborn violinplot() function</h5>

<li>This function draws a combination of boxplot and kernel density estimate.</li>
<li>A violin plot plays a similar role as a box and whisker plot.</li>
<li>It shows the distribution of quantitative data across several levels of one (or more) categorical variables such that those distributions can be compared.</li>
<li>Unlike a box plot, in which all of the plot components correspond to actual datapoints, the violin plot features a kernel density estimation of the underlying distribution.</li>
<li>I will plot the violinplot of Potential variable as follows-</li>

<h5>Multi-plot grids</h5>

<p>Seaborn FacetGrid() function¶</p>

<li>The FacetGrid class is useful when you want to visualize the distribution of a variable or the relationship between multiple variables separately within subsets of your dataset.</li>
<li>A FacetGrid can be drawn with up to three dimensions - row, col and hue. The first two have obvious correspondence with the resulting array of axes - the hue variable is a third dimension along a depth axis, where different levels are plotted with different colors.</li>
<li>The class is used by initializing a FacetGrid object with a dataframe and the names of the variables that will form the row, column or hue dimensions of the grid</li>
<li>These variables should be categorical or discrete, and then the data at each level of the variable will be used for a facet along that axis.</li>

```python
g=sns.FacetGrid(fifa19data,col="Preferred Foot")
```

<p>We can draw a univariate plot of Potential variable on each facet as follows-</p>

```python
g=sns.FacetGrid(fifa19data,col="Preferred Foot")
g=g.map(plt.hist,"Potential")
```

```python
g=sns.FacetGrid(fifa19data,col="Preferred Foot")
g=g.map(plt.hist,"Potential",bins=10,color="r")
```
